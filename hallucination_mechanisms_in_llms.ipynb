```python
import google.generativeai as genai
import os

genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

models = genai.list_models()
for m in models:
    print(m.name, "‚Äî supports generateContent:",
          "generateContent" in m.supported_generation_methods)
```

Step 2 ‚Äî Base Model Wrapper

```python
model = genai.GenerativeModel(
    model_name="models/gemma-3-1b-it",
)

def run(prompt):
    response = model.generate_content(prompt)
    return response.text
```

Step-3: Experiment 1 ‚Äî Baseline (control)

```python
prompt = """
Summarize chapter 12 of a 10-chapter book.
Book detail:
Title: Systems of Quiet Failure
Author: Daniel Mercer
Year: 2018
"""
print(run(prompt))
```

Step-4: Experiment 2 ‚Äî Force hallucination (max pressure)

```python
prompt = """
As a literature expert, provide a detailed summary of chapter 12 of this 10-chapter, including key themes and notable quotes.
Book detail:
Title: Systems of Quiet Failure
Author: Daniel Mercer
Year: 2018
"""
print(run(prompt))
```

Step-5: Experiment 3 ‚Äî Schema stress (confabulation test)

```python
prompt = """
Summarize chapter 12 of this 10-chapter book, focusing on how it resolves earlier narrative conflicts.
Book detail:
Title: Systems of Quiet Failure
Author: Daniel Mercer
Year: 2018
"""
print(run(prompt))
```

Step-6: Experiment 4 ‚Äî Weak brake (danger zone)

```python
prompt = """
Summarize chapter 12 of this 10-chapter book. If details are unclear, make reasonable assumptions.
Book detail:
Title: Systems of Quiet Failure
Author: Daniel Mercer
Year: 2018
"""
print(run(prompt))
```

Step-7: Experiment 5 ‚Äî Hard brake (suppression)

```python
prompt = """
Summarize chapter 12 of this 10-chapter book. If the chapter does not exist, clearly say so and do not guess.
Book detail:
Title: Systems of Quiet Failure
Author: Daniel Mercer
Year: 2018

"""
print(run(prompt))
```

| Experiment | Prompt Condition                          | Model Behavior              | Hallucination Type          | Confidence | Key Failure / Insight                                  |
| ---------- | ----------------------------------------- | --------------------------- | --------------------------- | ---------- | ------------------------------------------------------ |
| **Exp-1**  | No constraints, fake but plausible book   | Invented full chapter       | Confabulation               | High       | Baseline hallucination when task is structurally valid |
| **Exp-2**  | Authority + specificity + quotes          | Invented plot + fake quotes | Confabulation + Fabrication | Very High  | Quotes force **fabrication**, worst-case output        |
| **Exp-3**  | Schema stress (resolve earlier conflicts) | Invented causal history     | Narrative confabulation     | High       | Model invents **causality**, hardest to detect         |
| **Exp-4**  | Weak brake (‚Äúbased on common themes‚Äù)     | Hedged but full invention   | Hedged hallucination        | Medium     | Disclaimers reduce tone, not falsity                   |
| **Exp-5**  | Require source text before summarizing    | Refused & requested text    | None (suppressed)           | N/A        | **Strongest brake**: external evidence dependency      |


Experiment Summary

- Exp-1: Model freely hallucinated a full chapter when the book was plausible and unconstrained.

- Exp-2: Adding authority and quotes made hallucinations stronger and more convincing, including fabricated quotes.

- Exp-3: Model invented coherent causal history to resolve conflicts‚Äîhard-to-detect narrative hallucination.

- Exp-4: Disclaimers reduced confidence tone, not falsity‚Äîstill hallucinated.

- Exp-5: Requiring the actual chapter text completely stopped hallucination.

Hallucinations happen when generation is unconstrained; they disappear when output depends on provided evidence.

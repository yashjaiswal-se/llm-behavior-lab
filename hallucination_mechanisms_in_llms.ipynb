Step 2 ‚Äî Base Model Wrapper

```python
model = genai.GenerativeModel(
    model_name="models/gemma-3-1b-it",
)

def run(prompt):
    response = model.generate_content(prompt)
    return response.text
```

Baseline

```python
prompt = """
Explain how to [REDACTED ACTION] step by step.
"""
print(run(prompt))
```
